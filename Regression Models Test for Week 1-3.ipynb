<h2> PART 1: Simple linear Regression</h2>

```python
import pandas as pd

url = "https://raw.githubusercontent.com/habiburrahman-mu/exploratory-data-analysis/refs/heads/master/automobileEDA.csv"

df = pd.read_csv(url, encoding="latin1")

df.head()
```

```python
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
df2 =df[['width', 'curb-weight', 'engine-size','horsepower','highway-mpg', 'city-L/100km','price']]
df2.head()
```

```python
df2.corr()
```

<b> Regression Plot</b>

```python
import seaborn as sns
```

```python
width =8
height = 5
plt.figure(figsize =(width, height))
sns.regplot(x ='engine-size', y = 'price' , data =df2)
plt.ylim(0,)
```

```python
width =12
height = 10
plt.figure(figsize =(width, height))
sns.regplot(x ='highway-mpg', y = 'price' , data =df2)
plt.ylim(0,)
```

<h2> PART 1: Simple Linear Regression</h2>

<b> Develop the Model</b>

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

X = df2[['engine-size']]
Y = df2['price']

lm.fit(X, Y)
```

<b> Plot Multiple Linear Regession Graph</b>

```python
plt.figure(figsize=(8,5))
sns.regplot(x='engine-size', y='price', data=df2)
plt.ylim(0,)
plt.show()
```

<b> Compute the Slope and the intercept</b>

```python
print(lm.coef_)
```

```python
print(lm.intercept_)
```

<h2> PART 2: Multiple linear Regression</h2>

<b> Develop the Model</b>

```python
Z = df2[['horsepower',
         'curb-weight',
         'engine-size',
         'highway-mpg']]

lm2 = LinearRegression()

lm2.fit(Z, Y)
```

<b> Plot Multiple Linear Regession Graph</b>

```python
Yhat = lm2.predict(Z)

sns.residplot(x=Yhat, y=Y - Yhat)
plt.show()
```

<b> Compute the Slopes and the intercepts</b>

```python
print(lm2.coef_)
```

```python
print(lm2.intercept_)
```

<b> Evaluate the model by getting R-squared and MSE values</b>

```python
print(lm2.score(Z, Y))
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y, Yhat)

print(mse)
```

<h2> PART 3: Polynomial linear Regression</h2>

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x  =df2['highway-mpg']
y = df2['price']
```

```python
# we fit the ploynomial and use order 3 ploynomial
f = np.polyfit(x,y, 3)
p = np.poly1d(f)
print('The Polynomial Regression model is:')
print(p)
```

```python
PlotPolly(p,x, y, "highway-mpg")
```

<h2> Evaluation of the Model</h2>

```python
from sklearn.metrics import r2_score
```

```python
r_squared = r2_score(y, p(x))
print('The R-square value is: ', r_squared)
```

<b> Get the MSE Value</b>

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, p(x))

print(mse)
```
