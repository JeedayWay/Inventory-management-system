```python
# Question 1A

def add_book(catalog, book_id, title, author):
    book = {
        "id": book_id,
        "title": title,
        "author": author,
        "available": True
    }

    catalog.append(book)
    return catalog


def remove_book(catalog, book_id):

    for book in catalog:

        if book["id"] == book_id:

            if book["available"] == False:
                print("Cannot remove borrowed book")
                return catalog

            catalog.remove(book)
            print("Book removed successfully")
            return catalog

    print("Book not found")
    return catalog


catalog = [
    {
        "id": 101,
        "title": "The Great Gatsby",
        "author": "F. Scott Fitzgerald",
        "available": True
    }
]

catalog = add_book(
    catalog,
    102,
    "Python Basics",
    "John Smith"
)

print("Updated Catalog")

for book in catalog:
    print(book)

remove_book(catalog, 102)
```

```python
# Question 1B

class BorrowingHistory:

    def __init__(self):
        self.stack = []

    def borrow_book(self, book_id):
        self.stack.append(book_id)
        print(f"Book {book_id} borrowed")

    def return_book(self):

        if len(self.stack) > 0:
            return self.stack.pop()

        return None

    def view_last_borrowed(self):

        if len(self.stack) > 0:
            return self.stack[-1]

        return None


history = BorrowingHistory()

history.borrow_book(101)
history.borrow_book(102)

print("\nLast borrowed book:", history.view_last_borrowed())

print("Returned book:", history.return_book())
```

```python
# Question 1C

class Waitlist:

    def __init__(self):
        self.queue = []

    def add_to_waitlist(self, user_name, book_id):

        self.queue.append(
            {
                "user": user_name,
                "book_id": book_id
            }
        )

    def notify_next_user(self, book_id):

        for person in self.queue:

            if person["book_id"] == book_id:
                self.queue.remove(person)
                return person["user"]

        return None

    def view_waitlist(self, book_id):

        waiting = []

        for person in self.queue:

            if person["book_id"] == book_id:
                waiting.append(person["user"])

        return waiting


waitlist = Waitlist()

waitlist.add_to_waitlist("Alice", 101)
waitlist.add_to_waitlist("Brian", 101)
waitlist.add_to_waitlist("Carol", 101)

print("Current Waitlist:")
print(waitlist.view_waitlist(101))

print("\nNext user notified:",
      waitlist.notify_next_user(101))
```

```python
# Question 2A

menu = [
    ("Burger", 8.99, 15),
    ("Pizza", 12.99, 20),
    ("Salad", 6.99, 10),
    ("Pasta", 10.99, 18)
]


def find_expensive_items(menu, threshold):

    return [
        item
        for item in menu
        if item[1] > threshold
    ]


def sort_by_preparation_time(menu):

    return sorted(
        menu,
        key=lambda item: item[2]
    )


print("Expensive Items:")
print(find_expensive_items(menu, 10))

print("\nSorted Menu:")
print(sort_by_preparation_time(menu))
```

```python
# Question 2B

class OrderStack:

    def __init__(self):

        self.orders = []
        self.cancelled = []

    def place_order(self, order):

        self.orders.append(order)

    def cancel_last_order(self):

        if len(self.orders) > 0:

            order = self.orders.pop()

            self.cancelled.append(order)

            return order

        return None

    def undo_cancellation(self):

        if len(self.cancelled) == 0:

            print("No order to restore")
            return

        order = self.cancelled.pop()

        self.orders.append(order)

        return order


stack = OrderStack()

stack.place_order(
    {
        "order_id": 1,
        "items": ["Burger"]
    }
)

stack.place_order(
    {
        "order_id": 2,
        "items": ["Pizza"]
    }
)

print("Cancelled Order:")
print(stack.cancel_last_order())

print("\nRestored Order:")
print(stack.undo_cancellation())
```

```python
# Question 2C

class KitchenQueue:

    def __init__(self):

        self.queue = []

    def add_order(self, order_id, items):

        self.queue.append(
            {
                "order_id": order_id,
                "items": items
            }
        )

    def process_next_order(self):

        if len(self.queue) > 0:
            return self.queue.pop(0)

        return None

    def view_pending_orders(self):

        return self.queue

    def get_queue_length(self):

        return len(self.queue)

    def rush_order(self, order_id, items):

        self.queue.insert(
            0,
            {
                "order_id": order_id,
                "items": items
            }
        )


kitchen = KitchenQueue()

kitchen.add_order(101, ["Burger"])
kitchen.add_order(102, ["Pizza"])
kitchen.add_order(103, ["Pasta"])

print("Pending Orders:")
print(kitchen.view_pending_orders())

print("\nProcessed Order:")
print(kitchen.process_next_order())

print("\nQueue Length:")
print(kitchen.get_queue_length())
```
