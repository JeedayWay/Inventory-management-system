```python
from collections import deque

print("System Initialized")
```

```python
papers = []

papers.append("AI in Healthcare")
papers.append("Machine Learning Basics")

print(papers)
```

```python
paper_details = {}

paper_details[101] = {
    "title": "AI in Healthcare",
    "author": "Morgan",
    "year": 2026
}

print(paper_details)
```

```python
undo_stack = []

undo_stack.append("Deleted Paper 101")
undo_stack.append("Updated Paper 102")

print("Before undo:", undo_stack)

undo_stack.pop()

print("After undo:", undo_stack)
```

```python
submission_queue = deque()

submission_queue.append("Paper A")
submission_queue.append("Paper B")

print("Before processing:", submission_queue)

submission_queue.popleft()

print("After processing:", submission_queue)
```
